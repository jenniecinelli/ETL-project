## Extracting data using Pandas

```python
forbes_path = '../Resources/forbes_billionaires_geo.csv'
forbes_df = pd.read_csv(forbes_path)
forbes_df.head()
```

```python
bank_path = '../Resources/2015 World Bank data by nation and region.xls'
world_bank_df = pd.read_excel(bank_path)
world_bank_df.head()
```

## Cleaning Data

```python
bill_df = forbes_df[["Name", "NetWorth", "Country", "Residence", "Status", "Children", "Education", "Self_made" ]].copy()
```

```python
cities = []
states = []
res = bill_df["Residence"]
country = bill_df.Country

for idx, r in enumerate(res):
    try:
        (city, state) = r.split(", ")
        cities.append(city)
        
        if country[idx] == "United States":
            states.append(state)
        else:
            states.append("")
            

    except:
        cities.append("")
        states.append("")
    
```

```python
bill_df["city"]= cities
bill_df["state"] = states
bill_df = bill_df.drop("Residence",axis=1)
```

```python
ed = bill_df["Education"]
```

```python
dropOuts = []
school1 = []
school2 = []

for x in ed:
    try:
        if x.find(";") > 0:
            sch1, sch2 = x.split(';')
            deg1,sch1 = sch1.split(', ')
            deg2,sch2 = sch2.split(', ')
            dropOuts.append(deg1 == "Drop Out" | deg2 ==  "Drop Out")
            school1.append(sch1)
            school2.append(sch2)
        else:
            deg, sch1 = x.split(', ')
            dropped = deg == "Drop Out"
            dropOuts.append(dropped)
            school1.append(sch1)
            school2.append("")
    except:
        dropOuts.append(False)
        school1.append("")
        school2.append("")
            
        
```

```python
data = {"Drop Out":dropOuts,"School1": school1, "School2": school2}

df_4 = pd.DataFrame(data)
```

```python
df_bill = pd.concat([bill_df,df_4], axis=1)
df_bill = df_bill.drop("Education",axis=1)
```

```python
df_bill
```

```python
df_bill = df_bill.rename(columns = {"Name":"name",
                         "NetWorth": "net_worth",
                         "Country": "country_name",
                          "Self_made":"self_made",
                          "Drop Out": "drop_out"
                         })
```

```python
df_bill2 = df_bill[["name", "net_worth", "country_name", "city", "state", "Status", "Children", "drop_out", "School1","School2", "self_made"]].copy()
```
