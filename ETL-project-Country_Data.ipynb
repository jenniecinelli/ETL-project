```python
import pandas as pd
```

## Extracting data using Pandas

```python
bank_path = 'Resources/2015 World Bank data by nation and region.xls'
world_bank_df = pd.read_excel(bank_path)
world_bank_df.head()
```

## Cleaning Data

```python
world_bank_df.rename(columns={"Country Name":"country_name","Country Code":"country_code",
                              "Region Code":"region_code", 
                              "GDP, PPP (current international $)":"gdp",
                              "Population, total":"population", 
                              "Population CGR 1960-2015":"population_cgr", 
                              "Internet users (per 100 people)":"internet_users",
                              "2014 Life expectancy at birth, total (years)":"life_expectancy"}, inplace=True)
```

```python
world_bank_df.head(100)
```

```python
.drop(columns={"Popltn Largest City % of Urban Pop", 
                             "Literacy rate, adult female (% of females ages 15 and above)",
                            "Exports of goods and services (% of GDP)"}, inplace=True)
```

```python
country_name = world_bank_df.country_name
region_code = world_bank_df.region_code
country_code = world_bank_df.country_code
pop = world_bank_df.population
gdp = world_bank_df.gdp
internet_users = world_bank_df.internet_users
population_cgr = world_bank_df.population_cgr
life_expectancy = world_bank_df.life_expectancy
```

```python
population = []
for item in pop:
    try:
        new_item = int(item)
        population.append(new_item)
        
    except:
        new_item = ""
        population.append(new_item)
```

```python
gdp_ = []
for item in gdp:
    try:
        new_item = item/1000000
        gdp_.append(int(new_item))
        
    except:
        new_item = ""
        gdp_.append(new_item)
```

```python
internet = []
for item in internet_users:
    try:
        new_item = int(item)
        internet.append(new_item)
        
    except:
        new_item = ""
        internet.append(new_item)
```

```python
cgr = []
for item in population_cgr:
    try:
        new_item = round(100*item,2)
        cgr.append(new_item)
        
    except:
        new_item = ""
```

```python
life = []
for item in life_expectancy:
    try:
        new_item = int(item)
        life.append(new_item)
        
    except:
        new_item = ""
        life.append(new_item)
```

```python
world_bank_final_df = pd.DataFrame({"country_name":country_name, "country_code": country_code, "region_code": region_code, 
                                           "gdp": gdp_, "population": population, "population_cgr": cgr,
                                           "internet_users": internet, "life_expectancy": life})
```

```python
world_bank_final_df.head(100)
```

```python
world_bank_final_df.to_csv("world_bank_data.csv")
```
